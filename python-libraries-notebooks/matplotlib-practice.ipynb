```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
x = np.arange(0,10)
y = np.arange(11,21)
plt.scatter(x,y,c='b')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph in 2D")
plt.savefig('Test.png')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D Fig')
plt.savefig('Test1.png')
```

```python
plt.subplot(2,2,1) # subplots
plt.plot(x,y,'b*')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'y-')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
